```python
# the modules are imported and assigned an alias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# the dataset is imported and assigned to the variable data
data = pd.read_csv('gun-violence-data_01-2013_03-2018.csv')

# column 'n_killed' is assigned to variable x to be represented on the x axis.
x = data['n_killed']
# column 'n_injured' is assigned to variable y to be represented on the y axis.
y = data['n_injured']

# plt.scatter function is called and passed the data to correlate the columns.
plt.scatter(x, y) 
plt.show()
```

```python
# the correlation coefficient is printed
print(np.corrcoef(x, y))
plt.scatter(x, y) 

# add titles to the plot and axes
plt.title('A plot to show the correlation between n_killed and n_injured')
plt.xlabel('n_killed')
plt.ylabel('n_injured')

# add the line of best fit 
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='red')
plt.show()
```
